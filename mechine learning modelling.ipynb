```python
from sklearn.preprocessing import LabelEncoder
```

```python
le= LabelEncoder()
object_c=marg.select_dtypes(include='object').columns
for i in object_c:
    marg[i]=le.fit_transform(marg[i])
```

```python
X=marg.drop(['high_risk_applicant'],axis=1)
y=marg.iloc[:,11]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
list_algo=[LogisticRegression(),DecisionTreeClassifier(),RandomForestClassifier(),KNeighborsClassifier(),SVC(),GaussianNB()]
```

```python
 for algo in list_algo:
    model=algo
    model.fit(X_train,y_train)
    y_prepd=model.predict(X_test)
    print(f'the accuracy score for {algo} Regression Model is {round((algo.score(X_test,y_test)*100),2)}','%')
```

```python
Lclassifier = LogisticRegression(random_state=10)
Lclassifier.fit(X_train,y_train)
y_pred=Lclassifier.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
